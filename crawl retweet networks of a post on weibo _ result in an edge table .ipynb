```python
import requests
from pprint import pprint
import time
import csv
import random  # generate a random number
import pandas as pd  # generate data frame
from time import sleep  # stop the loop and have a rest
import re
pattern = re.compile(r'(?<=//@).*?(?=:)')
```

```python
headers = {
    "User-Agent": "***",
    "Referer":"https://weibo.com/5605235135/Kq5i4pTom",
    "Cookie":"***",
    "X-Requested-With": "XMLHttpRequest"
}
```

```python
def parse(id,page):
    
    for i in range(1,int(page)+1):
        #爬取的url
        if i == 1:
            url = "https://weibo.com/ajax/statuses/repostTimeline?id={}&page={}&moduleID=feed&count=10".format(id,i)
        else:
            url = "https://weibo.com/ajax/statuses/repostTimeline?id={}&page={}&moduleID=feed&count=20".format(id,i)
        
        #获取页面数据
        res=requests.get(url,headers=headers)
        data = res.json()['data']
        #pprint(data)
        
        #爬取具体node信息
        for info in data:
            #print(info)
            raw_text = info['text_raw']
            #print(raw_text)
            if "//@" in raw_text:
                try:
                    from_node = pattern.search(raw_text)
                    from_node_name = from_node.group()
                except:
                    from_node_name = ' '
                    print("有转发，但无法获取node",url,raw_text)
            else:
                from_node_name = "NCP生命支援"   #手动改成博主昵称
                
            try:
                to_node_name = (info['user']['screen_name'])
                create = info['created_at']
            except:
                to_node_name = ' '
                create = ' '
            
            from_node_list.append(from_node_name)
            to_node_list.append(to_node_name)
            create_time.append(create)

            #print(from_node_name,to_node_name)
        print("%d页获取完成"%i)
        time.sleep(1)
```

```python
#eadgetable
from_node_list = []
to_node_list = []
create_time = []
parse(4661177742795029,1000)  #传入参数,改id
```

```python
print(from_node_list,to_node_list,create_time)
```

```python
edge_dict = {
    'Source':from_node_list,
    'Target':to_node_list,
    'Time':create_time
}
```

```python
edgedata = pd.DataFrame(edge_dict)
edgedata
```

```python
edgedata.to_csv('edge_table_1_1001_大v.csv')   #修改文件名字 edge_table_{id}_{转发数}.csv
```
